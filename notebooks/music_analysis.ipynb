```python
def get_playlist_items(playlist_id):
    results = sp.user_playlist_tracks(playlist_id=playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks

def get_saved_items():
    results = sp.current_user_saved_tracks(limit=50)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks

#playlist_items = get_playlist_items(playlist_id)
playlist_items = get_saved_items()
```

```python
#print(playlist_tracks[0].keys())
#print(playlist_tracks[0]['track'].keys())
#print(playlist_tracks[0]['track']['album'].keys())

"""
Audio analysis : https://developer.spotify.com/documentation/web-api/reference/#/operations/get-audio-analysis 
Audio features : https://developer.spotify.com/documentation/web-api/reference/#/operations/get-several-audio-features
"""
def playlist_items_to_df(playlist_items):
    tracks = []
    for track in playlist_items:
        t = track['track']
        
        """
        try:
            features = sp.audio_analysis(t['id'])
        except spotipy.SpotifyException:
            track = dict()
            tracks.append(track)
            continue
        """
            
        track = {
            'id': t['id'],
            'name': t['name'],
            'artist': ', '.join([artist['name'] for artist in t['artists']]),
            'duration_ms': t['duration_ms'],
            'explicit': t['explicit'],
            'release_date': t['album']['release_date'],
            'popularity': t['popularity'],
        }

        tracks.append(track)
    df = pd.DataFrame(tracks)
    return df

df = playlist_items_to_df(playlist_items)
df
```

```python
batch_size = 100
all_features = pd.DataFrame()
for i in range(df.shape[0] // 100 + 1):
    index_min = i * batch_size
    index_max = (i+1) * batch_size
    print(index_min, index_max)
    track_ids = df['id'][index_min:index_max].values
    features = pd.DataFrame(sp.audio_features(tracks=track_ids))
    all_features = pd.concat([all_features, features])

print(all_features.shape)
all_features
```

```python
for col in ['danceability', 'energy', 'speechiness', 'mode', 'tempo', 'time_signature', 'valence']:
    print(col)
    #df = df.drop(col, axis=1)
    df[col] = all_features[col].values
    print(df[col].describe())
    print()
```
